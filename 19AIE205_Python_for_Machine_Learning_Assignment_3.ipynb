**Overfitting and Underfitting**

Building a model that is too complex for the amount of
information we have is called **overfitting**.

Building a model that is too simple, may not be able to capture all aspects. This is called **underfitting**

While we can make up many rules that work well on available data, remember that we are not interested in making predictions

# 1. MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
X=np.array([[1,2,3,4,5,6,500],[-100,-50,0,50,100,2000,10000],[2,1,1,1,1,1,1]]) #3 Samples, 7 attributes or features
print(X)
print(X.shape)
scaler = MinMaxScaler()
scaler.fit(X)
xscale=scaler.transform(X)
print(xscale)
```

**Min Max Scaling**
$$xscale=\frac{x-x_{min}}{x_{max}-x_{max}}$$
**Example**

How 0.99019608 is obtained?

Values in first columns are $1,-100,2$. $x_{min}=-100$,$x_{max}=2$

Calculating xscale for first value, that is $x=1$
$$xscale=\frac{1-(-100)}{2-(-100)}$$
which means
$$xscale=\frac{101}{102}$$
