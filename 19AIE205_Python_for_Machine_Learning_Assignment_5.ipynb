# 2. Steps in Clustering


1.   Preprocess Data (Removing or Replacing or Scaling )
2.   Create Similarity Metric (Ex: Eucledian Distance)
3. Run Clustering Algorithm (Stopping criteria can be number of iterations)
4. Interpret Results (Visualizing using matplotlib or other libraries)



# 3. K-Means Clustering
To group data into $k$ clusters, K-Means perform as follows,

**Steps**


1.   Randomly chose centroid for each cluster
2.   Assign each sample to closest centroid based on Similarity Metric (Eucledian Distance)
3. For each cluster, recompute centroid, by taking average of all samples in that cluster
4. Repeat 2 and 3 until stopping criteria is met (Ex : 50 iterations)



# 4. Simple Numerical Example
K-Means.PNG

The example shows the result after iteration $1$. 

This process will be repeated till stopping criteria is met

#5. Implementing the simple example in Python

```python
from sklearn.cluster import KMeans
import numpy as np
X = np.array([[1, 2], [1, 4], [5, 3],[5, 6]]) #Samples
start1=np.array([[1,2],[1,4]]) #Centroid-1
kmeans = KMeans(n_clusters=2,max_iter=1,init=start1).fit(X)  #Defining and fitting K-means

print(kmeans.labels_) #Will print Cluster allocation of each samples
print(kmeans.predict([[0, 0], [12, 3]])) #Predicting for unseen data
print(kmeans.cluster_centers_) #Centroids after iteration 1


```

## Visualizing results after iteration 1

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(6, 6))
plt.scatter(X[kmeans.labels_ == 0, 0], X[kmeans.labels_ == 0, 1], c='green', label='cluster 1')
plt.scatter(X[kmeans.labels_ == 1, 0], X[kmeans.labels_ == 1, 1], c='blue', label='cluster 2')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='*', s=300,  c='r', label='centroid')
plt.legend()
```

```python
#Converting to Dataframe
import pandas as pd
import io
df = pd.read_csv(io.BytesIO(uploaded['faithful.csv']))
```

```python
#Visualizing Input Data
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 6))
plt.scatter(df.iloc[:, 0], df.iloc[:, 1])
plt.xlabel('Eruption time in mins')
plt.ylabel('Waiting time to next eruption')
plt.title('Visualization of raw data');
```

```python
#K-Means Clustering
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Standardize the data
X_std = StandardScaler().fit_transform(df)

# Run local implementation of kmeans
km = KMeans(n_clusters=2)
km.fit(X_std)
centroids = km.cluster_centers_

# Plot the clustered data
fig, ax = plt.subplots(figsize=(6, 6))
plt.scatter(X_std[km.labels_ == 0, 0], X_std[km.labels_ == 0, 1],c='green', label='cluster 1')
plt.scatter(X_std[km.labels_ == 1, 0], X_std[km.labels_ == 1, 1],c='blue', label='cluster 2')
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], marker='*', s=300, c='r', label='centroid')
plt.legend()
plt.xlim([-2, 2])
plt.ylim([-2, 2])
plt.xlabel('Eruption time in mins')
plt.ylabel('Waiting time to next eruption')
plt.title('Visualization of clustered data', fontweight='bold')
ax.set_aspect('equal');
```
